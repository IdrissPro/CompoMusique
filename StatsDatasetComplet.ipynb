```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)

DataSet=pd.read_csv('dataset_16notes.csv')
X=DataSet.copy()
```

```python
print(X.describe())
```

```python
plt.figure(figsize=(30,30))  


for column in X.columns:
    plt.subplot(16,3, X.columns.get_loc(column) + 1)  
    value_counts = X['note_10_1'].value_counts()
    plt.bar(value_counts.index, value_counts.values)
    plt.title(column)

plt.tight_layout()
plt.show()
```
