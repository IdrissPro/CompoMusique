```python
import pandas as pd
import matplotlib.pyplot as plt
# Charger les données depuis le fichier CSV
nom_fichier = "Suivi entrainement LSTMusic.csv"
donnees = pd.read_csv(nom_fichier)
# Récupérer les colonnes pertinentes
neurones = donnees['nombre de neurones'].values
minMSE = donnees['minMSE'].values
batch = donnees['batch Normalization'].values
drop = donnees['Dropout'].values
pourcent = donnees['Pourcentage du dataset'].values
epoch = donnees['epoch'].values
```

```python
# Tracer le nuage de points avec des noms
plt.figure(figsize=(4, 8))
# Utiliser plt.scatter pour tracer un nuage de points

plt.scatter(pourcent, minMSE, color='b', label='minMSE')
# Ajouter des noms à chaque point
for i in range(len(minMSE)):
    if batch[i] and drop[i]:
        plt.annotate("Dropout,BatchNorm,"+str(neurones[i])+"neurones", (pourcent[i], minMSE[i]))
    elif batch[i]:
        plt.annotate("BatchNorm,"+str(neurones[i])+"neurones", (pourcent[i], minMSE[i]))
    elif drop[i]:
        plt.annotate("Dropout,"+str(neurones[i])+"neurones", (pourcent[i], minMSE[i]))
    else:
        plt.annotate(str(neurones[i])+"neurones", (pourcent[i], minMSE[i]))
# Ajouter des étiquettes et un titre
plt.xlabel('Pourcentage du Dataset')
plt.ylabel('MSE minimal atteint sur le jeu de test')
#plt.title('Nuage de points avec noms')
# Afficher la légende
plt.xticks([1,10, 100])
plt.legend()
# Afficher le nuage de points
plt.show()
```
