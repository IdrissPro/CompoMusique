```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)

DataSet10=pd.read_csv('MINI_10_dataset_16notes.csv')
Dataset1=pd.read_csv('MINI_1_dataset_16notes.csv')

X10,X1=DataSet10.copy(),Dataset1.copy()
print(X1['note_10_1'].value_counts())
```

```python
print(X1.describe())
```

```python
print(X10.describe())
```

```python
plt.figure(figsize=(30,30))  


for i,column in enumerate(X1.columns,1): #enumerate(X.columns,1) pour commencer  1
    plt.subplot(16,3,i)
    value_counts = X1[column].value_counts()
    plt.bar(value_counts.index, value_counts.values)
    if(i%3==0 or i%3==2):
        plt.xlim(0,100)        #s'arrter avant 100 pour l'axe des abscisses des dures 
    plt.title(column)

plt.tight_layout()
plt.show()
```
